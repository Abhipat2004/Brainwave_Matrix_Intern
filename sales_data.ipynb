```python
from google.colab import files
import pandas as pd
import io

# file upload
uploaded = files.upload()

# Automatic filename
filename = next(iter(uploaded))

# change encoding
df = pd.read_csv(io.BytesIO(uploaded[filename]), encoding='ISO-8859-1')

# Display first few rows
df.head()
```

```python
# View columns and data types
df.info()

#null values check
df.isnull().sum()
```

```python
df.describe()
```

```python
# Check columns
df.columns

# Drop non-useful columns
df.drop(['Row ID', 'Postal Code'], axis=1, inplace=True)
```

```python
# Total Sales and Profit by Category
category_sales = df.groupby('Category')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
category_sales
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#plot style
sns.set(style="whitegrid")
```

```python
category_sales.plot(kind='bar', figsize=(10, 5), title='Sales and Profit by Category')
plt.ylabel('Amount')
plt.show()
```

```python
region_data = df.groupby('Region')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
region_data
```

```python
top_subcategories = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).head(10)
top_subcategories.plot(kind='barh', figsize=(10,6), title='Top 10 Sub-Categories by Profit')
plt.xlabel('Profit')
plt.show()
```
